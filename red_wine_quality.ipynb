## Red Wine Quality

There is a red winequality dataset which is quite famous. This dataset has 11 different attributes that  decide the quality of red wine.  

These attributes include:
- fixed acidity
- volatile acidity
- citric acid
- residual sugar
- chlorides
- free sulfur dioxide
- total sulfur dioxide
- density
- pH
- sulphates
- alcohol  

Based on these different attributes, we are required to predict the quality of red wine  which is a value between 0 and 10. 

How to solve the *overfitting* for this dataset.

We can treat this problem either as a classification problem or as a regression  problem since wine quality is nothing but a real number between 0 and 10. For  simplicity, letâ€™s choose classification.

> Reference: Thakur, Abhishek. Approaching (Almost) Any Machine Learning Problem. 

```python
import pandas as pd

# https://www.kaggle.com/uciml/red-wine-quality-cortez-et-al-2009?select=winequality-red.csv
df = pd.read_csv('winequality-red.csv')
```

```python
df.head(10)
```

```python
# mapping dictionary that maps the quality values from 0 to 5 
quality_mapping = {
    3:0,
    4:1,
    5:2,
    6:3,
    7:4,
    8:5
}

# use the map function of pandas to convert the values in a given column
df.loc[:, 'quality'] = df.quality.map(quality_mapping)
```
